```

from scipy.stats import spearmanr
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
import pandas as pd
```

```
 from google.colab import files
from google.colab import drive
drive.mount('drive')
```

```
results = pd.read_csv("drive/My Drive/Correlations_results.csv")
```

```
print(results)
```

```
#Statement coverage results 
statement_coverage_results = results["statement"]
#Branch coverage test results 
branch_coverage_results = results["branch"]
#Mutation coverage test results
mutation_coverage_results = results["mutation"]
#adaptive
adaptive_coverage_results = results["adaptive"]
#bugs
bugs_coverage_results = results["bugs"]
```

```
print(bugs_coverage_results)
```

```
print(spearmanr(statement_coverage_results, mutation_coverage_results))
```

```
print(pearsonr(statement_coverage_results, mutation_coverage_results))
```

```
mutation_coverage_g,statement_coverage_results_g = zip(*sorted(zip(mutation_coverage_results,statement_coverage_results)))
plt.plot(mutation_coverage_g, statement_coverage_results_g)
plt.ylabel("Statement coverage results")
plt.xlabel("Mutation Coverage results")
plt.title("Statement coverage vs Mutation Coverage")
plt.show()
```

```
print(spearmanr(branch_coverage_results, mutation_coverage_results))
```

```
print(pearsonr(branch_coverage_results, mutation_coverage_results))
```

```
mutation_coverage_g,branch_coverage_results_g = zip(*sorted(zip(mutation_coverage_results,branch_coverage_results)))
plt.plot(mutation_coverage_g, branch_coverage_results_g)
plt.ylabel("Branch coverage results")
plt.xlabel("Mutation Coverage results")
plt.title("Branch coverage vs Mutation Coverage")
plt.show()
```

```
bugs_coverage_g,statement_coverage_results_g = zip(*sorted(zip(bugs_coverage_results,statement_coverage_results)))
plt.plot(adaptive_coverage_g, statement_coverage_results_g)
plt.ylabel("Statement coverage results")
plt.xlabel("Bugs Coverage results")
plt.title("Statement coverage vs Bugs Coverage")
plt.show()
```

```
bugs_coverage_g,branch_coverage_results_g = zip(*sorted(zip(bugs_coverage_results,branch_coverage_results)))
plt.plot(adaptive_coverage_g, branch_coverage_results_g)
plt.ylabel("Branch coverage results")
plt.xlabel("Bugs Coverage results")
plt.title("Branch coverage vs Bugs Coverage")
plt.show()
```
